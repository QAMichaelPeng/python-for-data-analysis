```python
from lxml import objectify
with open('../examples/mta_perf/Performance_MNR.xml') as f:
    parsed = objectify.parse(f)
```

```python
root = parsed.getroot()
data=[]
skip_fields=['PARENT_SEQ', 'INDICATOR_SEQ', 'DESIRED_CHANGE', 'DECIMAL_PLACES']
for elt in root.INDICATOR:
    el_data = {}
    for child in elt.getchildren():
        if child.tag in skip_fields:
            continue
        el_data[child.tag]=child.pyval
    data.append(el_data)
data
```

```python
perf=DataFrame(data)
perf
```

```python
perf.head()
```

## SQL

```python
import sqlite3
```

```python
query="""
create table test(
a VARCHAR(20),
b VARCHAR(20),
c REAL,
d INTEGER
)
"""
conn = sqlite3.connect("mydata.sqlite")
```

```python
conn.execute(query)
```

```python
conn.commit()
```

```python
data=[
    ('WA', 'Hello', 1.3, 3),
    ('CA', 'World', 5.0, 4),
    ('NY', 'City', 3.3, 9)
]
stmt= 'INSERT INTO TEST VALUES(?,?,?,?)'
conn.executemany(stmt, data)
```

```python
conn.commit()
```

```python
cursor = conn.execute('select * from test')
rows=cursor.fetchall()
rows
```

```python
cursor.description
```

```python
df=DataFrame(rows, columns=[x[0] for x in cursor.description])
```

```python
df
```
