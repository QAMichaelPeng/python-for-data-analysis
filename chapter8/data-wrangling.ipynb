## Union and concat

```python
df1=DataFrame({'key':list('bbacaab'), 'data1':range(7)})
df1
```

```python
df2=DataFrame({'key':list('abd'), 'data2':range(3)})
df2
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, on='key')
```

```python
pd.merge(df1,df2,left_on='key',right_on='key')
```

```python
pd.merge(df1,df2,left_on='key',right_on='key',how='outer')
```

```python
pd.merge(df1,df2,left_on='key',right_on='key',how='left')
```

```python
df3=DataFrame({'key':list('bbbaaaddd'), 'data2':range(9)})
df3
```

```python
pd.merge(df1, df3)
```

### Concatenation

```python
arr=np.random.randint(20, size=(3,4))
arr
```

```python
np.concatenate([arr,arr])
```

```python
np.concatenate([arr,arr],axis=1)
```

```python
s1=Series([0,1],index=list("ab"))
s1
```

```python
s2=Series([2,3,4],index=list('cde'))
s2
```

```python
s3=Series([5,6],index=list('fg'))
s3
```

```python
pd.concat([s1,s2,s3])
```

```python
pd.concat([s1,s2,s3],axis=1,join='outer')
```

```python
s4=pd.concat([s1,s2])
s4
```

```python
pd.concat([s1,s4],axis=1)
```

```python
pd.concat([s1,s4],axis=1,join='inner')
```

```python
pd.concat([s1,s4],axis=1,join_axes=[['a','c','b','f']])
```

```python
result=pd.concat([s1,s2,s3],keys=['one','two','three'])
result
```

```python
result.unstack()
```

```python
result=pd.concat([s1,s2,s3],axis=1,keys=['one','two','three'])
result
```
